```python
def char_counts(s):
    count_chart = {}
    for i in s:
        if i in count_chart:
            count_chart[i] += 1
        else:
            count_chart[i] = 1
    return count_chart

input_string = input("Write something:" ).strip()
char_counts(input_string)


```

```python
def word_freq_count(s):
    words = s.split(" ")
    count = {}
    for i in words:
        if i in count:
            count[i] += 1
        else:
            count[i] = 1
    return count

sentence = input_string = input("Write something:" ).strip()
word_freq_count(sentence)
```
