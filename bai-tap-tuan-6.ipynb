```python
numb_input = input("Enter your numbers (e.g. 2 3 13): ")

def a1(input_numbs):
    numb_list = list(map(int, input_numbs.split()))
    res = []
    counted = []
    
    for i in range(len(numb_list)):
        if numb_list[i] not in counted:
            res.append(numb_list[i])
            counted.append(numb_list[i])
    return res


a1(numb_input)
```

```python
numb_input = input("Enter your numbers (e.g. 2 3 13): ")


def a2(input_numbs):
    numb_list = list(map(int, input_numbs.split()))
    res = []

    for i in range(len(numb_list)):
        if i == 0:
            res.append(numb_list[i])
        else:
            temp_sum = res[i-1] + numb_list[i]
            res.append(temp_sum)
    return res


a2(numb_input)
```

```python
def a4():
    input_v1 = input("").split()
    input_v2 = []
    my_dict = {}
    
    for i in input_v1:
        input_v2.append(tuple(i.split(":")))
    
    for i in range(len(input_v2)):
        key, value = input_v2[i]
        if key in my_dict:
            my_dict[key].append(value)
        else:
            my_dict[key] = [value]
    return my_dict

a4()

```

```python
def a5():
    numb_list = list(map(int, input("Enter your intergers: ").split(", ")))
    count = {
        "negatives" : 0,
        "positives" : 0,
        "zeros" : 0
    }
    
    for i in numb_list:
        if i < 0:
            count["negatives"] += 1
        elif i > 0:
            count["positives"] += 1
        else:
            count["zeros"] += 1

    return count

a5()
```

```python
def a8(string_input):
    count = {}
    for i in string_input:
        if i in count:
            count[i] += 1
        else:
            count[i] = 1
    return count
        
    
string_input = input().split(", ")
a8(string_input)
```

```python
def add_dict_to_dict(my_dict, target):
    for i in my_dict:
        if i in target:
            target[i] += my_dict[i]
        else:
            target[i] = my_dict[i]
    return target

def a9(dict1, dict2):
    new_dict = {}
    add_dict_to_dict(dict1, new_dict)
    add_dict_to_dict(dict2, new_dict)

    new_dict = dict(sorted(new_dict.items()))
    return new_dict

dict1 = {'a': 100, 'b': 200, 'c': 300}
dict2 = {'a': 300, 'b': 200, 'd': 400}

a9(dict1, dict2)
```

```python
def a10(numb_list, k):
    result = {}
    for i in numb_list:
        if i not in result and i not in result.values():
            temp = k - i
            if temp in numb_list:
                result[min(temp, i)] = max(i, temp)
    

    return result.items()

numb_list = list(map(int, input("Enter a list of numbers: ").split()))
k = int(input("Enter the sum target: "))

a10(numb_list, k)
```
